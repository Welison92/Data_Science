Pronto! Já identificamos as variáveis do nosso conjunto de dados e seus tipos. No entanto, encontramos um pequeno problema nas colunas que contêm datas. As datas estão em um formato não convencional. Em um tópico posterior chamado 'Formatação dos dados', iremos padronizar as datas para um formato comumente utilizado em um **DataFrame**. Essa padronização é importante para facilitar análises futuras.




# Formatação dos dados

```python
def formatar_data(lista_datas):
    datas_formatadas = []  # Cria uma lista vazia para armazenar as datas formatadas
    for data in lista_datas:  # Itera sobre cada elemento da lista de datas
        if isinstance(data, float):  # Verifica se o elemento é do tipo float
            datas_formatadas.append(np.nan)  # Se for, adiciona NaN à lista de datas formatadas
        else:
            data_formatada = datetime.strptime(str(data)[:8], "%Y%m%d")  # Converte a data para o formato desejado (YYYYMMDD)
            datas_formatadas.append(data_formatada)  # Adiciona a data formatada à lista de datas formatadas

    return datas_formatadas  # Retorna a lista de datas formatadas


colunas_datas = [
    'Data Protocolo',
    'Data CTPS Gerada',
    'Data Emissão',
    'Data Nascimento',
]

for nome_coluna in colunas_datas:  # Itera sobre cada nome de coluna na lista de colunas de datas
    datas = dados_CTPS_2020_2022[nome_coluna].tolist()  # Obtém a lista de datas da coluna específica
    dados_CTPS_2020_2022[nome_coluna] = formatar_data(datas)  # Chama a função formatar_data para formatar as datas e
    # substitui a coluna original pelos valores formatados
```

# Resumo estatístico
Infelizmente, não é possível realizar um resumo estatístico das variáveis numéricas neste dataset, pois todas ou quase todas as colunas contêm variáveis categóricas. Um resumo estatístico é uma análise descritiva que envolve medidas como média, desvio padrão, mínimo, máximo e quartis, que são aplicáveis apenas a variáveis numéricas. No entanto, para variáveis categóricas, podemos realizar análises diferentes, como contagem de frequência, proporções e gráficos de barras para visualizar a distribuição dos valores nas categorias.

# Análise exploratória de dados
Começaremos a nossa análise verificando a distribuição da nossa variável alvo **Tipo Protocolo**.

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(7, 4))
sns.countplot(data=dados_CTPS_2020_2022, x='Tipo Protocolo')
plt.xlabel('Tipo de Protocolo')
plt.ylabel('Número de ocorrências')
plt.title('Distribuição dos tipos de protocolo')
plt.show()
```

Até o momento a única informação que temos é que existe uma maior concentração de protocolos de 1ª via.
